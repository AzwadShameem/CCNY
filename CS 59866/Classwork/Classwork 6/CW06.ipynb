```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
x1s = np.linspace(-3, 3, 100)
plt.plot(x1s)
```

```python
x1s = np.linspace(-3, 3, 100)
w = np.random.uniform(size=(3, 2))
bias = np.random.uniform(size=(1, 2))[0]
ans = []


def h(x, w, bias):
    # implement neural network
    z0_0 = w[0][0]*bias[0] + w[0][1]*x
    z0_1 = w[1][0]*bias[1] + w[1][1]*x
    a0_0, a0_1 = max(0, z0_0), max(0, z0_1)
    z1_0 = w[2][0]*a0_0 + w[2][1]*a0_1
    return 1/(1+np.exp(-z1_0))

for i in x1s:
    ans.append(h(i, w, bias))
plt.plot(ans)
```

```python
# Plotting 16 different weights/biass
figure, axis = plt.subplots(4, 4)
for i in range(4):
    for j in range(4):
        ans = []
        w = np.random.uniform(size=(3, 2))
        for k in x1s:
            ans.append(h(k, w, bias))
        axis[i, j].plot(ans)
```
